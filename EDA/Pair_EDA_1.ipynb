# Ejercicio 2 
Cread dos DataFrames nuevos, uno que incluya solo las variables categóricas y otro que incluya solo las variables numéricas.


```python
df_numericas = df.select_dtypes(include = ['float64'])
df_numericas.head()
```

```python
df_categoricas = df.select_dtypes(include = ['object'])
df_categoricas.head()
```

- ¿Cuáles son los principales estádisticos de los DataFrames?

```python
df_numericas.describe()
```

Vemos que para las variables numéricas los principales estadísticos son contar valores, la media, desviación típica, valor mínimo, los cuartiles 1 y 3 (25% y 75%), y el valor máximo.

```python
df_categoricas.describe()
```

Como vimos en el ejercicio anterior si usamos describe() incluyendo todas las columnas, si tenemos muchas variables categóricas, habrá un gran número de Nan. Por lo que para verlo todo más claro es mejor separar por tipo de variable y observar sus propiedades en tablas más pequeñas y adaptadas a la información que queremos obtener.

- Extraed los valores únicos. 

```python
lista_unicos = []
for col in df_categoricas:
    val_unicos = df_categoricas[col].unique()
    lista_unicos.append(val_unicos)
```

```python
lista_unicos
```

```python
variable = df_categoricas.columns.unique()
variable
```

BONUS presenta los resultados en un DataFrame. 

```python
diccionario = {'variable':variable, 'n_unique':lista_unicos}
diccionario
```

```python
pd_df = pd.DataFrame(diccionario)
pd_df
```
