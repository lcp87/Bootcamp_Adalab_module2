```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
import pandas as pd
```

```python
archivo_csv = pd.read_csv('odisea.csv')
archivo_csv.head(5)
```

```python
df = pd.DataFrame(archivo_csv)
```

```python
columnas = df[['type', 'country']]
columnas
```

```python
eeuu = columnas[(columnas['country'] == 'USA')]
eeuu
```

```python
num_eeuu = eeuu['type'].value_counts()
df_num_eeuu = (num_eeuu).reset_index()
df_num_eeuu
```

```python
y1 = df_num_eeuu['type']
x1 = df_num_eeuu['index']
```

```python
españa = columnas[(columnas['country'] == 'SPAIN')]
españa
```

```python
num_españa = españa['type'].value_counts()
df_num_españa = (num_españa).reset_index()
df_num_españa
```

```python
y2 = df_num_españa['type']
x2 = df_num_españa['index']

x2.size
```

```python
# ataques en USA y España, cada subplot corresponderá a un país
fig,ax = plt.subplots(1,2,figsize=(15,5))  #indice

#grafica 1
ax[0].set_xlabel(xlabel='Tipos de ataques')
ax[0].set_ylabel(ylabel='Nº de ataques')
ax[0].plot(x1, y1, c = 'black', marker = 'o')

# #grafica 2
ax[1].set_xlabel(xlabel='Tipos de ataques')
ax[1].set_ylabel(ylabel='Nº de ataques')
ax[1].plot(x2, y2, c = 'black', marker = 'o')

# #titulo
ax[0].set(title ='Tipos de ataques en USA')
ax[1].set(title = 'Tipos de ataques en España')

plt.show()
```
