```python
# 7. Extraed los ataques que se hayan producido en el año 2000, 1997 y 2006. Utilizad el método isin() para resolver este ejercicio.
df_ataques_2000 = df[df["year"].isin([2000.0])]
df_ataques_2000.head(2)
```

```python
# ataques que han ocurrido por tiburón blanco 
# Usad el método str.contains() para resolver este ejercicio.

df_ataque_tiburon = df[df['species_'].str.contains('white', 'White', na= False, regex = True)]
df_ataque_tiburon.head(2)
```

```python
patron = 'W|w[hite]|T|t[iger]'
```

```python
# 9. Extraed ahora los ataques por tiburón blanco o tigre. Utilizad el método str.contains() para resolver este ejercicio.
df_tiburon_tigre = df[df['species_'].str.contains(patron, na= False, regex = True)]
df_tiburon_tigre.sample(5)
```

```python
# 10 . Utilizando la columna de year cread una nueva columna llamada siglo que siglo-XX o siglo-XXI. 
# Donde el año sea menor que 2000 pondremos siglo-XX y el resto de los años siglo-XXI.

df['siglo'] = np.where(df['year'] >= 2000.0, 'siglo-XXI', 'siglo-XX')
df.head(5)
```

```python
orden_columnas = [df.columns]
orden_columnas

nuevo_orden_columnas = ['case_number', 'year', 'siglo', 'type', 'country', 'area', 'location',
        'activity', 'name', 'sex_', 'age', 'time', 'species_', 'href']

df = df.reindex(columns = nuevo_orden_columnas)
df.head(5)
```
