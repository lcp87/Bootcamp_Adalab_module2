```python
# porcentaje de valores nulos. ######## ELimina prints

casos_nulos = archivo_df.isnull().sum()
print(casos_nulos)

casos_totales = archivo_df.shape[0]         


resultado = (casos_nulos * 100) / casos_totales
print(resultado)


# % -->
#  casos total ----- 100 %
#  casos que tengo --- x % --->  X = (casos nulos * 100%) / casos totales
```

```python
# Sacad los valores únicos y su frecuencia de las variables numéricas. 
# Almacenad los resultados en un diccionario.             

df_numericas = archivo_df[["Year", "original order"]]
dic = {}

for i in df_numericas:
    df_numericas[i].value_counts()
    dic[i] = df_numericas[i].value_counts()
```

```python
#cambiando los nombres de cols a minúsculas
#rename (utiliza dic. Keys: nombre antiguo, value nombre neuvo)

nuevas_columnas = {}

for col in archivo_df.columns:
    nuevas_columnas[col] = col.lower()
print(nuevas_columnas)
```

```python
# ejemplo de Ana de list comprehension
# lista = ["Laura", "Almudena"]

# lista_nueva = []


# for nombre in lista:
#     lista_nueva.append(len(nombre))
# lista_nueva

# lista_nueva2 = [len(nombre) for nombre in lista]
# lista_nueva2
```

```python
# conviertiendo nombres cols a minúsculas con list cromprehension

nuevas_columnas2 = { col: col.lower() for col in archivo_df.columns}
nuevas_columnas2
```

```python
# cambiando el nombre de las columnas
archivo_df.rename(columns = nuevas_columnas, inplace = True)
archivo_df.head()
```

```python
# reemplazad espacios por _

nuevas_columnas3 = {}

for col in archivo_df.columns:
    nuevas_columnas3[col] = col.replace(" ", "_")

archivo_df.rename(columns = nuevas_columnas3, inplace = True)
archivo_df.head()
```

```python
## reemplazad espacios por _ con dict comprehension
chorizo = {col: col.replace(" ", "_") for col in archivo_df.columns}
```

```python
# Eliminad las columnas de Unnamed: 22 y Unnamed: 23, Case Number.1, Case Number.2, original order, 
# Investigator or Source, pdf, href formula.

archivo_df.drop(["unnamed:_22", "unnamed:_23", "case_number.1", "case_number.2", "original_order", 
"investigator_or_source", "pdf", "href_formula"], axis = 1, inplace= True)
```
