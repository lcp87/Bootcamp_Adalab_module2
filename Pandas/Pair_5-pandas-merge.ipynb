```python
import pandas as pd
```

```python
# 1.Carga los tres ficheros nombrados anteriormente y los exploramos para familizarizarnos con ellos 
# y saber que información tienen en común y debatir con tu compañera como podríamos juntar toda la información.
historico = pd.read_csv("attacks_historico.csv", index_col = 0)
historico.head(2)
```

```python
adicional = pd.read_csv("attacks_adicional.csv", index_col = 0)                     #desaparece la columna unnamed_0.
adicional.head(2)
```

```python
odisea = pd.read_csv("odisea.csv")                                                  #odisea: archivo de csv proveniente de PairPandas3
odisea.head(2)
```

```python
# Unificamos los nombres de las columnas

#cambiamos el nombre en historico
nuevas_columnas = {}

for col in historico.columns:
    nuevas_columnas[col] = col.lower()
    historico.rename(columns = nuevas_columnas, inplace = True)
    nuevas_columnas[col] = col.replace(" ", "_")
print(historico)
```

```python
#cambiamos el nombre en adicional
for col in adicional.columns:
    nuevas_columnas[col] = col.lower()
    adicional.rename(columns = nuevas_columnas, inplace = True)
    nuevas_columnas[col] = col.replace(" ", "_")
print(adicional)
```

```python
# Modificar la df 'historico' para que sea igual a 'odisea'


#observamos cómo son nuestos dfs
print(odisea.shape)
print(historico.shape)
```

```python
#observamos historico para ver qué columnas eliminar
historico.head(2)
```

```python
#eliminamos las cols que no nos interesan
historico.drop(["unnamed:_22", "unnamed: 23", "case_number.1", "case_number.2", "original_order", 
"investigator_or_source", "pdf", "href_formula"], axis = 1, inplace= True)
```

```python
#comparamos las columnas de las dos dfs que queremos unir: historico y odisea
historico.columns
```

```python
odisea.columns
```

```python
# cambiamos el orden de las columnas de historico al mismo que de odisea (el csv de pandas 3)
historico.reindex(columns= odisea.columns).head(2)
```

```python
# 4. Unir dfs ya iguales (historico y odisea)

histodisea = pd.concat([historico, odisea], axis = 0, join = 'outer', ignore_index = False)
```

```python
#comprobamos que la unión de las dos dfs (histodisea) ha salido bien.
odisea.shape
```

```python
# 4. Unir histodisea + adicional = final

final = histodisea.merge(adicional, how = "left", on = "case_number")
final.head(2)

# el mejor método es merge porque los nombres de nuestras columnas son diferentes (descartado concat) 
# y porque no tenemos una columna índice (join descartado)
```

```python
# Guardar archivo en csv
final.to_csv("definitivo.csv")
```
