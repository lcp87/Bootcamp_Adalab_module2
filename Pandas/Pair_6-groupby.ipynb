```python
# Es el momento de poner en pr√°ctica los groupbyde pandas. Para esto vamos a trabajar con el csv creado en el ejercicio de pair programming anterior.
# El objetivo de este ejercicio es agrupar los datos para contestar las siguientes preguntas:
```

```python
import pandas as pd
```

```python
definitivo = pd.read_csv('definitivo.csv', index_col= 0)
definitivo.head(5)
```

```python
# ¬øCu√°ntos ataques hubo por pa√≠s? Devuelve un dataframe *****
ataque_pais = definitivo.groupby('country')['case_number'].count()
df_ataque_pais = pd.DataFrame(ataque_pais)
df_ataque_pais
```

```python
# üìå NOTA Ordena los resultados de mayor a menor. Esto todav√≠a no lo hemos visto, pero un dataframe lo podemos ordenar con .sort_values().
#  Aqu√≠ ten√©is una explicaci√≥n de como hacerlo.
ataques_ordenados = ataque_pais.sort_values(ascending=False)
df_ataques_ordenados = pd.DataFrame(ataques_ordenados)
df_ataques_ordenados
```

```python
# ¬øCu√°ntos ataques ha habido por a√±o? Devuelve un dataframe. De nuevo ordena los resultados de mayor a menor
ataque_a√±o = definitivo.groupby('year')['case_number'].count()
df_ataque_a√±o = pd.DataFrame(ataque_a√±o)
df_ataque_a√±o
```

```python
# ordena los resultados de mayor a menor
a√±o_ordenado = ataque_a√±o.sort_values(ascending=False)
df_a√±o_ordenado = pd.DataFrame(a√±o_ordenado)
df_a√±o_ordenado
```

```python
# ¬øCu√°ntos ataques hubo por a√±o y sexo? Devolved un dataframe solo del a√±o 2000 en adelante.
ataque_a√±o_sexo = pd.DataFrame(definitivo.groupby(['year','sex_'])['case_number'].count()).reset_index()
ataque_a√±o_sexo
```

```python
# Filtrado a partir del a√±o 2000
filtrado_a√±o_sexo = ataque_a√±o_sexo[ataque_a√±o_sexo['year'] > 2000.0]
filtrado_a√±o_sexo.head(5)
```

```python
# 4. Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. ¬øCu√°ntos de los ataques fueron letales por sexo? Devuelve un dataframe.
pais_fatal = definitivo.groupby(['country', 'sex_', 'fatal_(y/n)'])['case_number'].count().reset_index()
pais_fatal
```

```python
filtrado_pais = pais_fatal[pais_fatal['country'] == 'USA']
filtrado_pais.head(5)
```
