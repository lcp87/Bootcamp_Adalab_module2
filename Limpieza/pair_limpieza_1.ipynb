## Ejercicio 4
- Cambiar a minuscula en la columna 'country'

```python
def minusculas(df):
    for indice in range(len(df)):
        try:
            df[indice] = df[indice].lower()
        except:
            pass

    
minusculas(df_copia['country'])
```
